```python
from sage.all import *
import networkx as nx
import matplotlib.pyplot as plt
from math import *
from itertools import *
```

```python
D = WeylCharacterRing("A4").dynkin_diagram(); D
```

```python
W =  WeylGroup(["B", Integer(3)])
s1, s2, s3 = W.simple_reflections()
a = W.one() ; b = s1*s2*s1*s3 
G = W.bruhat_graph(a, b)

Gx = nx.DiGraph()          
for e in G.edges():
    Gx.add_edge(e[0], e[1])

POS = {} ; L = []
for k in range(b.length()+1):
    L.append([x for x in W.elements_of_length(k) if x in G.vertices()])

for i, Lk in enumerate(L):
    M = len(Lk)
    for k, elem in enumerate(Lk):
        POS[elem] = (k - M/2, i)
        
options = {
    'node_color': 'magenta',
    'node_size':25,
    'edge_color' : 'magenta',
    'width': 1,
}
plt.title('Bruhat interval [a,b] in '+ str(W))
nx.draw(Gx, POS, with_labels=True, **options)
plt.draw()
```
